49. Group Anagrams

Medium

https://leetcode.com/problems/group-anagrams/description/

Given an array of strings strs, group the 
anagrams
 together. You can return the answer in any order.

 

Example 1:

Input: strs = ["eat","tea","tan","ate","nat","bat"]

Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

Explanation:

There is no string in strs that can be rearranged to form "bat".
The strings "nat" and "tan" are anagrams as they can be rearranged to form each other.
The strings "ate", "eat", and "tea" are anagrams as they can be rearranged to form each other.
Example 2:

Input: strs = [""]

Output: [[""]]

Example 3:

Input: strs = ["a"]

Output: [["a"]]

 

Constraints:

1 <= strs.length <= 104
0 <= strs[i].length <= 100
strs[i] consists of lowercase English letters.

```python
#  sol Optimised
from typing import List

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        print('optimised Solution')
        anagram_map = {}
        for s in strs:
            s_sorted = ''.join(sorted(s))
            if s_sorted in anagram_map:
                anagram_map[s_sorted].append(s)
            else:
                anagram_map[s_sorted] = [s]
        
        return list(anagram_map.values())
```

```python
#  sol 1
from typing import List

class Solution:
    def get_car_map_str(self, s: str) -> str:
        car_map_s = [0]*26
        for c in s:
            car_map_s[(ord(c)-97)] += 1 
        return ','.join([str(i) for i in car_map_s])
        
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagram_map = {}
        for s in strs:
            s_map = self.get_car_map_str(s)
            if s_map in anagram_map.keys():
                anagram_map[s_map].append(s)
            else:
                anagram_map[s_map] = [s]
        
        result = list(anagram_map.values())
        result.reverse()# for Test case matching , Not required While submitting solution
        
        for i in range(len(result)):# for Test case matching , Not required While submitting solution
            result[i].sort()
        
        # print(result)
        return result
             
        
```
